## Idea: Analyzing incoming Review or text data to understand Brand Sentiment

```python
import pandas as pd

# Loading file into Dataframe 

df = pd.read_csv('hotel_reviews.csv')

# making sure our column in question is a string for TextBlob

df['reviewtext']=df['reviewtext'].astype(str) 
df
```

```python
# Assigning a polarity score to the comment using textblob

from textblob import TextBlob
polarity=[]
for sentence in df['reviewtext']:
    analysis = TextBlob(sentence)
    polarity.append(analysis.sentiment.polarity)

# Adding polarity as a column in the dataframe

df['polarity']=polarity

# Adding sentiment, based on polarity, as very positive, positive, somewhat positive, somewhat negative, negative,
# very negative or neutral

sentiment=[]
for value in polarity:
    if (value > 0.3):
        x = 'Positive'
    elif (0.1 < value < 0.3):
        x = 'Somewhat Positive'
    elif (value <= 0.1):
        x = 'Negative'
    else:
        x ='Neutral'
    sentiment.append(x)

# Adding sentiment as a column in the dataframe

df['sentiment'] = sentiment

df
```
